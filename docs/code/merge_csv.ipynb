```python
from pandas import Series, DataFrame
from PIL import Image
import pandas as pd
import csv
import glob
import os

#
# ------------------------------
#  Real     id        name
#   0  ==   1    ==   zero
#   1  ==   2    ==   one
#   2  ==   3    ==   two
#   3  ==   4    ==   three
#   4  ==   5    ==   four
#   5  ==   6    ==   five
#   6  ==   7    ==   six
#   7  ==   8    ==   seven
#   8  ==   9    ==   eight
#   9  ==  10    ==   nine
# ------------------------------
#

# train, test 경로 변경
os.chdir("/home/internship/anaconda3/envs/cyp\
/lib/python3.6/site-packages/tensorflow\
/models/research/object_detection/images")
path = '.'
files_in_dir = [f for f in os.listdir(path) if f.endswith('csv')]
first = True
column_name = ["class", "xmin", "ymin", "bbox_width", "bbox_height"]
for filenames in files_in_dir:
    df = pd.read_csv(filenames, names=column_name)
    image_name = filenames[:-3] + "png"
    
    im = Image.open(image_name)
    im_width, im_height = im.size
    
    
    df['filename'] = image_name  
    df['width'] = im_width
    df['height'] = im_height
    df['xmax'] = df['xmin'] + df['bbox_width']
    df['ymax'] = df['ymin'] + df['bbox_height']
    df = DataFrame(df, columns = ['filename', 'width', 'height', 'class','xmin','ymin','xmax', 'ymax'])
    df['class'] = df['class'].replace({0: 'zero', 1: 'one', 2: 'two', 3: 'three',
                                   4: 'four', 5: 'five', 6: 'six', 7: 'seven', 8: 'eight',
                                   9: 'nine'})

    # train_labels, test_labels
    if(first): 
        df.to_csv('train_labels.csv', mode='a', index=False)
        first = False
    else:
        df.to_csv('train_labels.csv', header=None, mode='a', index=False)
        
    
print("Done")
```
